```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import re
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('/content/tripadvisor_hotel_reviews.csv', engine="python")
df.head()
```

```python
conditions = [(df['Rating']<=2),
              (df['Rating']==3),
              (df['Rating']>3)
              ]
values = [0,1,2]
df['Rating']=np.select(conditions,values)
df.head(30)
  
```

```python
df['Review'] = df['Review'].str.lower()
for i in range(len(df['Review'])):
  df['Review'][i] = re.sub(r"http\S+", " ",df['Review'][i])
  df['Review'][i] = re.sub("[^a-zA-Z]+", " ",df['Review'][i])
  df['Review'][i] = re.sub(" +", " ",df['Review'][i])
df['Review'][4712]
```

```python
import nltk
for i in range(len(df['Review'])):
  df['Review'][i]=nltk.word_tokenize(df['Review'][i])
df['Review'][0]
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```

```python
stop = set(stopwords.words('english'))
for i in range(len(df['Review'])):
  df['Review'][i]=[ w for w in df['Review'][i] if w not in stop]
df['Review'][1]
```

```python
ps = PorterStemmer()
for i in range(len(df['Review'])):
  for j in range(len(df['Review'][i])):
    df['Review'][i][j]=ps.stem(df['Review'][i][j])
```

```python
df['Review'][1]
```

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer
for i in range(len(df['Review'])):
  df['Review'][i]=TreebankWordDetokenizer().detokenize(df['Review'][i])
```
